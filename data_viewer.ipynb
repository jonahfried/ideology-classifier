# Stitching together DFs, then filtering for complete rows and columns pertaining to the text and score.

```python
left_url_to_story.rename(index=str, columns={"story_id":"stories_id"}, inplace=True)
left_compound_info = left_url_ideos.join(
    left_url_to_story.set_index("url") , on='url'
).join(
    left_stories.set_index("stories_id"), on="stories_id"
)
compact_left_info = left_compound_info[["text", "ideo_est", "stddev_sharer_ideo", "stories_id"]]
compact_left_info = compact_left_info.dropna()
```

```python
right_url_to_story.rename(index=str, columns={"story_id":"stories_id"}, inplace=True)
right_compound_info = right_url_ideos.join(
    right_url_to_story.set_index("url") , on='url'
).join(
    right_stories.set_index("stories_id"), on="stories_id"
)
compact_right_info = right_compound_info[["text", "ideo_est", "stddev_sharer_ideo", "stories_id"]]
compact_right_info = compact_right_info.dropna()
# compact_right_info
```

## alt sampling with similar amounts of each ideo (find out if this is useful)

```python
TRAIN_PROPORTION = .9

compact_left_info.rename(columns={"stories_id":"id"}, inplace=True) #For clarity and generality
compact_left_info["label"] = "left"

compact_right_info.rename(columns={"stories_id":"id"}, inplace=True) #For clarity and generality
compact_right_info["label"] = "right"

TEST_PROPORTION = 1 - TRAIN_PROPORTION
rows_left = compact_left_info.shape[0]
rows_right = compact_right_info.shape[0]
min_rows = min(rows_left, rows_right)

train_rows = int(min_rows*TRAIN_PROPORTION)
left_train_data = compact_left_info.sample(n=train_rows)
right_train_data = compact_right_info.sample(n=train_rows)

test_rows = int(min_rows*TEST_PROPORTION)
left_data_unused = compact_left_info[~compact_left_info.id.isin(left_train_data)]
left_test_data = left_data_unused.sample(n=test_rows)
right_data_unused = compact_right_info[~compact_right_info.id.isin(right_train_data)]
right_test_data = right_data_unused.sample(n=test_rows)

training_data = pd.concat([left_train_data, right_train_data])
testing_data = pd.concat([left_test_data, right_test_data])
```
